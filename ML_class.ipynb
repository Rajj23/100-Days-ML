```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.linear_model as linear_model
```

```python
x= np.array([23,34,64,32,19,37,48]).reshape(-1,1)
y = np.array([63,56,35,66,83,42,73])
```

```python
model = linear_model.LinearRegression()
model.fit(x,y)
```

```python
y_pred = model.predict(x)
```

```python
plt.scatter(x,y,color='blue',label='Actual Data')
plt.plot(x,y_pred,color='red',label='Best Fit Line')
plt.xlabel('Car Age (Years)')
plt.ylabel('Y-axis')
plt.title('Car Speed (km/h)')
plt.legend()
plt.show()
```

```python
data = pd.DataFrame({'num':[78,85,90,88,76]})
```

```python
df = pd.DataFrame({'Study_Hours':[2,3,5,7,8,4,6,9,10,1],
                      'Exam_Score':[50,60,75,85,90,65,80,95,98,45]})
df.sample(2)
```

```python
print("Correlation Coefficient: ",df.corr().iloc[0,1])
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df['Study_Hours'],y=df['Exam_Score'])
plt.title("Study Hours vs Exam Score")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.show()
```

Boxplot (Numerical vs. Categorical)

```python
plt.figure(figsize=(4,4))
sns.boxplot(x=pd.cut(df['Study_Hours'],bins=3, labels=['Low','Medium','High']),y=df['Exam_Score'])
plt.title("Boxplot of Study Hours vs Exam Score")
plt.show()
```
